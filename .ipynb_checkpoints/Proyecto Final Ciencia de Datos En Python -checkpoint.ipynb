## Definition of linear regression model function:

```python
def linear_regression(x,y, epochs, learning_rate, print_error_each):
    
    x_matrix = np.array([x, np.ones_like(x)]).transpose() ##design matrix. contains x and 1
    beta = np.array([129,30000]) ## initial parameters
    ##empty array used for storing errors
    error_iterations = np.array([]) 
    model = np.array([])
    beta0 = np.array([])
    beta1 = np.array([])
    for i in range(0,epochs+1):
        y_hat = np.matmul(x_matrix,beta)
        error_matrix = (y - y_hat)**2
        error = np.mean(error_matrix)*0.5
        error_iterations = np.append(error_iterations, error)
        
        ##### gradients
        gradient = np.matmul((y_hat - y).transpose(), x_matrix)/len(x_matrix)
        m = beta[0] - learning_rate*gradient[0]
        b = beta[1] - learning_rate*gradient[1]
        beta[0] = beta[0] - learning_rate*gradient[0]
        beta[1] = beta[1] - learning_rate*gradient[1]
        beta0 = np.append(beta0,m)
        beta1 = np.append(beta1,b)
        model = np.append(model,beta)
        if i % print_error_each == 0:
            print("Error in iteration: ", i,":",error)
    model = np.vstack((beta0,beta1,error_iterations))
    return(model)
```

## Definition of Error graph function

```python

def error_graph(x,iteration):
    it = np.array([])
    error = np.array([])
    for i in range (1,len(x)):
        
        if i % iteration == 0: 
            it = np.append(it,i)
           # it[1] = np.array(i)
            error = np.append(error, x[i])
            
    graph = np.append(it,error)
    plt.figure(figsize=(6,6))
    plt.plot(it,error, color = "red")
    plt.scatter(it,error, color = "orange", alpha = 0.5)
    plt.xlabel('Iteration')
    plt.ylabel('Error')
    plt.title('Error graph')
    plt.show()
    
    
        
```

## Third Model 

```python
def average_model(first_model,second_model, x):
    first_model_prediction = first_model[0,-1]*x + first_model[1,-1]
    second_model_prediction = second_model.predict(x.reshape(-1,1))
    third_model_prediction = (first_model_prediction + second_model_prediction)/2
    
    return(first_model_prediction,second_model_prediction,third_model_prediction)
    
    
```

```python
models_1= average_model(model_1, reg_1, ds_test[:,1])
models_2= average_model(model_2, reg_2, ds_test[:,2])
models_3= average_model(model_3, reg_3, ds_test[:,3])
models_4= average_model(model_4, reg_4, ds_test[:,4])
models_5= average_model(model_5, reg_5, np.nan_to_num(ds_test[:,5]))
```

## Validating each model with test data

A function that computes and graphs the sum of squared errors for each model is defined.

```python
def sum_test_error_graph(models_x):
    first_model_error = np.sum((models_x[0] - ds_test[:,0]))**2
    second_model_error = np.sum((models_x[1] - ds_test[:,0]))**2
    third_model_error = np.sum((models_x[2] - ds_test[:,0]))**2
    
    errors = np.stack((first_model_error,second_model_error,third_model_error))
    
    bars = ('#1 Model', '#2 Model', '#3 Model')
    y_pos = np.arange(len(bars))
    plt.figure(figsize=(7,7))
    plt.bar(y_pos, errors)
    plt.title('Error Comparison Chart')
    plt.ylabel('Sum of squared errors')
    # Create names
    plt.xticks(y_pos, bars)
    plt.show()

    
    
   
```
